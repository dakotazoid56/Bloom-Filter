```python
import matplotlib.pyplot as plt
import csv
import sys
import math
```

```python
def read_csv_data(fileName):
    xValues = []
    yValues = []
    with open(fileName, mode='r', newline='') as file:
        csvReader = csv.reader(file)
        for row in csvReader:
            xValues.append(float(row[0]))
            yValues.append(float(row[1]))
    return xValues, yValues
```

```python
# Plot 1

#Change to take in two csv files, one in firstHash.csv and one in secondHash.csv
#Plot Both On the Same Graph but different color lines
x1Values, y1Values = read_csv_data('firstHash.csv')
x2Values, y2Values = read_csv_data('secondHash.csv')

plt.plot(x1Values, y1Values, 'r-', label='First Hash')  
plt.plot(x2Values, y2Values, 'b-', label='Second Hash')
plt.xlabel('Bin Weight')  
plt.ylabel('Average # Bins at Weight') 
plt.title('First vs. Second Hash Function Random Input Distribution (100 Average)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot 2

#Change to take in two csv files, one in firstHash.csv and one in secondHash.csv
#Plot Both On the Same Graph but different color lines
x1Values, y1Values = read_csv_data('firstHash2.csv')
x2Values, y2Values = read_csv_data('secondHash2.csv')


plt.plot(x1Values, y1Values, 'r-', label='First Hash')  
plt.plot(x2Values, y2Values, 'b-', label='Second Hash')
plt.xlabel('Bin Weight')  
plt.ylabel('Average # Bins at Weight') 
plt.title('First vs. Second Hash Function Linear Input Distribution (100 Average)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot 3

def theoreticalFalsePositiveRate(c, k):
    return (1 - math.exp(-k/c))**k


x1Values, y1Values = read_csv_data('falsePositive10H1.csv')
x2Values, y2Values = read_csv_data('falsePositive10H2.csv')

x3Values = range (4,14)
y3Values = [theoreticalFalsePositiveRate(10, k) for k in x3Values]

kOptimal = round(10 * math.log(2))

plt.plot(x1Values, y1Values, 'r-', label='Hash 1')
plt.plot(x2Values, y2Values, 'b-', label='Hash 2')
plt.plot(x3Values, y3Values, 'g-', label='Theoretical')
plt.axvline(x=kOptimal, color='k', linestyle='--', label=f'Optimal k = {kOptimal}')

plt.xlabel('Number Hash Functions (k)')
plt.ylabel('False Positive Median')
plt.title('False Positives for K, c = 10')
plt.grid(True)
plt.legend()
plt.show()
```
